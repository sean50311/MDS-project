```python
# 1. 欄位名稱對應（僅要對應的欄位做改名）
col_map = {
    'Air temperature (K)': 'Air temperature',
    'Process temperature (K)': 'Process temperature',
    'Rotational speed (rpm)': 'Rotational speed',
    'Torque (Nm)': 'Torque',
    'Tool wear (min)': 'Tool wear',
}
```

```python
# 用 rename 只處理指定欄位
df = new_data.rename(columns=col_map).copy()
```

```python
# 2. 新增 failure 欄位
for col in ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']:
    df[col] = 0
```

```python
# 3. 解析 Diagnostic 欄位，自動 one-hot
failure_map = {
    'Tool Wear Failure': 'TWF',
    'Heat Dissipation Failure': 'HDF',
    'Power Failure': 'PWF',
    'Overstrain Failure': 'OSF',
    'Random Failure': 'RNF'
}
```

```python
def parse_failure(diag):
    # 支援多個failure（分號、逗號、空格等分隔）
    result = {k:0 for k in failure_map.values()}
    if pd.isnull(diag) or diag.strip() == '' or diag == 'No Failure':
        return result
    # 允許一列有多個failure
    for key, col in failure_map.items():
        if key in diag:
            result[col] = 1
    return result
```

```python
failures = df['Diagnostic'].apply(parse_failure).apply(pd.Series)
for col in ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']:
    df[col] = failures[col]
```

```python
# 4. 新增 Machine failure 欄位（有任一 failure 則為 1）
df['Machine failure'] = df[['TWF', 'HDF', 'PWF', 'OSF', 'RNF']].max(axis=1)
```

## 補值

```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
import os
```

```python
# 欄位名稱
tool_wear_col = 'Tool wear'
other_cols = [
    'Air temperature', 
    'Process temperature', 
    'Rotational speed', 
    'Torque'
]
```

```python
# 1. Tool wear 的補值方法（a）
def impute_toolwear_linear(df):
    df = df.copy()
    df[tool_wear_col] = df[tool_wear_col].interpolate(method='linear')
    return df

def impute_toolwear_rolling(df, window=5):
    df = df.copy()
    df[tool_wear_col] = (
        df[tool_wear_col]
        .fillna(df[tool_wear_col].rolling(window=5, min_periods=1).mean())
        .fillna(method='ffill')
        .fillna(method='bfill')
        .fillna(df[tool_wear_col].mean())
    )
    return df

def impute_toolwear_ffill(df):
    df = df.copy()
    df[tool_wear_col] = df[tool_wear_col].fillna(method='ffill').fillna(method='bfill')
    return df

toolwear_methods = {
    'linear': impute_toolwear_linear,
    'rolling': impute_toolwear_rolling,
    'ffill': impute_toolwear_ffill,
}
```

```python
# 2. 其他欄位的補值方法（b）
def impute_other_mean(df):
    df = df.copy()
    imputer = SimpleImputer(strategy='mean')
    df[other_cols] = imputer.fit_transform(df[other_cols])
    return df

def impute_other_median(df):
    df = df.copy()
    imputer = SimpleImputer(strategy='median')
    df[other_cols] = imputer.fit_transform(df[other_cols])
    return df

def impute_other_knn(df, n_neighbors=5):
    df = df.copy()
    imputer = KNNImputer(n_neighbors=n_neighbors)
    df[other_cols] = imputer.fit_transform(df[other_cols])
    return df

def imputed_other_rolling(df, window_size = 5):
    df = df.copy()
    for col in other_cols:
        df[col] = (
            df[col]
            .fillna(df[col].rolling(window=window_size, min_periods=1).mean())
            .fillna(method='ffill')
            .fillna(method='bfill')
            .fillna(df[col].mean())
        ) 
    return df

other_methods = {
    'mean': impute_other_mean,
    'median': impute_other_median,
    'knn': impute_other_knn,
    'rolling': imputed_other_rolling,
}
```

```python
root_dir = '../../instance/imputed_data'  

# 3. 批次組合補值
for t_name, t_func in toolwear_methods.items():
    for o_name, o_func in other_methods.items():

        df_tmp = df.copy()
        df_tmp = t_func(df_tmp)      # 補 Tool wear
        df_tmp = o_func(df_tmp)      # 補四個參數
        # 建立資料夾路徑（上層是 toolwear method，下層是 other method）
        out_dir = os.path.join(root_dir, t_name)
        os.makedirs(out_dir, exist_ok=True)
        # 給檔名
        filename = os.path.join(out_dir, f'PMDI_imputed_{t_name}_{o_name}.csv')
        df_tmp.to_csv(filename, index=False)
        print(f'輸出：{filename}')

print("全部補值組合已完成！")
```

## 論文方法 - 全用rolling average

```python
df = processed_new_data.copy()  # 或 pd.read_csv('你的檔案路徑')

cols = [
    'Air temperature',
    'Process temperature',
    'Rotational speed',
    'Torque',
    'Tool wear'
]

window = 5  # 可調整

for col in cols:
    # 先用 rolling average 補值
    df[col] = df[col].fillna(df[col].rolling(window=window, min_periods=1).mean())
    # 如果還有NaN（例如一開始多個NaN），可再補 ffill、bfill 或 mean
    df[col] = df[col].fillna(method='ffill').fillna(method='bfill').fillna(df[col].mean())


# 儲存
out_dir = os.path.join(root_dir, 'all-rolling')
os.makedirs(out_dir, exist_ok=True)
filename = os.path.join(out_dir, f'PMDI_imputed_all-rolling.csv')
df.to_csv(filename, index=False)
```
